```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('heart_disease_results.csv')
```

```python
df.head()
```

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
# loop over examples
for example in df["name"].unique():
    df_example = df[df["name"]==example]
    #print("Example: {}".format(example))
    
    # loop over test modality
    for modality in df_example["executionMode"].unique():
        df_mod = df_example[df_example["executionMode"]==modality]
        #print("Execution mode: {}".format(modality))
        
        unique_output_mode = df_mod["ouputMode"].unique()
        nrows = 1
        ncols = len(unique_output_mode)        
        fig, axes = plt.subplots(nrows=nrows, ncols=ncols)
        fig.set_size_inches(15, 5)
        fig.suptitle('{} - {}'.format(example, modality))
        
        for i in range(nrows*ncols):
            #for j in range(ncols):
            output_mode = unique_output_mode[i]
            #print("Output mode: {}".format(output_mode))
            df_output = df_mod[df_mod["ouputMode"]==output_mode]                    
                
            unique_dbms = df_output["dbms"].unique()
            for k in range(len(unique_dbms)):
                dbms = unique_dbms[k]
                #print("DBMS: {}".format(dbms))
                df_dbms = df_output[df_mod["dbms"]==dbms]
                axes[i].set_title(output_mode)
                #axes[i].set_xticks(df_dbms["batchSize"])
                df_dbms.plot(kind='line',x='batchSize',y='timeWithOutConnection',ax=axes[i], color=colors[k], label=dbms)
    
```
